#### Clean the data

Are there issues with the data? If so, can you solve these issues? 

Findings:

1) The initial check should be if data is basically a time-series. "isDataSorted" function helps user to understand if data is ordered and sort it otherwise. Though it may require further check, it is also obvious from the output of "df.head(10)" command that Dataset A requires a sort operation.

2) The second action is to detect outliers and convert them them into null values for interpolation. 

3) As seen from statistical summary, two data set have the same amount of data without any major issues in statistics.

```python
import numpy as np

def detectNull(data,colname): #detect if there are null values
    
    isnull = data[colname].isnull().values.any()
    if isnull:
        print(data.name + " contains null values")

def isDataSorted(data): #sort data if necessary
    isSorted = data["Time"].is_monotonic
    if not isSorted:
        data.sort_values(by="Time",inplace=True, ignore_index=True)
    return data

outliers = []
def detect_outliers_zscore(data):
    thres = 3
    mean = np.mean(data['Price'])
    std = np.std(data['Price'])

    for i in data['Price']:
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return data
    
#detect if there exists null values inside Price column
detectNull(dfA,'Price')
detectNull(dfB,'Price')

#Sort Data to construct a time-series
dfA = isDataSorted(dfA)
dfB = isDataSorted(dfB)

print("Dataset A")
print(dfA.head(10))
print("\n")
print("Dataset B")
print(dfB.head(10))

#Find outlier values in Dataset
dfA = detect_outliers_zscore(dfA)
dfB = detect_outliers_zscore(dfB)

locs = dfB.loc[dfB['Price'].isin(outliers)].copy()
for pos in locs.index:

    dfB['Price'].iloc[pos] = np.nan

#Describe the statistical features of data
statData = pd.DataFrame(describeData(dfA,'Price'))
statData.rename(columns={'Price':'PriceA'},inplace=True)
statData.insert(1,'PriceB',describeData(dfB,'Price'))
print(statData)
```

### 2 - Analyses

#### Correlation

Analyze the correlation between the two instruments. What does this say about the data and its potential uses in trading?

Findings:

1) Returns are assumed to be stationary, which will be verified by Augmented Dickey-Fuller test. The null hypothesis for ADF implies the data is non-stationary, therefore, if the test value is greater than critical value, it can be said that the data is stationary.

2) If returns are individually stationary, the correlation calculation can be made.

3) ADF Test results show that returns are I(0) and stationary. In addition, they are highly correlated according to three different correlation calculation, which implies that this pair, A & B, can be considered for a potential pairs trading, return to mean trading setup. As a straight-forward example, if the return of one asset among two experiences a shortfall on returns compared to other one, investor can enter into long position in the undervalued asset and short position in the overvalued asset assuming the historical correlation stays steady, which is a bold assumption. Therefore, more complex and hard-to-implement approach, cointegration that implies long-term relationship between prices, must be considered. 

```python
from statsmodels.tsa.stattools import adfuller

def ADFTest(data):
    
    res = adfuller(data,maxlag=1)

    # Printing the statistical result of the adfuller test
    print('Augmented Dickey_fuller Statistic: %f' % res[0])
    print('p-value: %f' % res[1])
 
    # printing the critical values at different alpha levels.
    print('critical values at different levels:')
    for k, v in res[4].items():
        print('\t%s: %.3f' % (k, v))

#Return  & Log Price calculation 
dfA['Returns'] = dfA['Price'].pct_change()
dfB['Returns'] = dfB['Price'].pct_change()
dfA['LogPrice'] = np.log(dfA['Price'])
dfB['LogPrice'] = np.log(dfB['Price'])

#For the sake of simplicity, another dataframe will be formed for the rest of analysis.
mergedData = pd.merge(dfA.iloc[:,5:7],dfB.iloc[:,5:7],left_index=True,right_index=True)
mergedData.rename(columns={'Price_x':'PriceA','Returns_x':'ReturnsA','Price_y':'PriceB','Returns_y':'ReturnsB'},inplace=True)

print('ADF Test for Return A')
ADFTest(mergedData['ReturnsA'].iloc[1:])
print('ADF Test for Return B')
ADFTest(mergedData['ReturnsB'].iloc[1:])

#Correlation between ReturnA and ReturnB
print("Correlation Values")
print('Pearsons R: :%f' % mergedData['ReturnsA'].corr(mergedData['ReturnsB']))
print('Spearmans Rho: %f' % mergedData['ReturnsA'].corr(mergedData['ReturnsB'],method='spearman'))
print('Kendalls Tau: %f' % mergedData['ReturnsA'].corr(mergedData['ReturnsB'],method='kendall'))

```

#### Cointegration

Analyze whether these instrument are cointegrated and report/explain your findings.

1) Before starting cointegration test, Enger-Granger, we need to make sure that price data is non-stationary. Though proving that prices are non-stationary time-series also means that log-prices are either(Alexander,2002), using log prices are the standard approach.

2) As seen from test results, both price data are non-stationary, which gives the opportunity to use Engle-Granger test for cointegration.

3) The test statistic of residual ADF test is less than critical value for even 1% (-3.52<-3.43). The null hypothesis, which is the data is non-stationary, can be rejected which proves that these instruments are cointegrated and suitable for pairs trading.

```python
#Check if Log Price is non-stationary.

mergedData['LogPriceA'] = np.log(mergedData['PriceA'])
mergedData['LogPriceB'] = np.log(mergedData['PriceB'])

print('ADF Test for Log Price A')
ADFTest(mergedData['LogPriceA'])
print('ADF Test for Log Price B')
ADFTest(mergedData['LogPriceB'])

#Cointegration test
import statsmodels.api as sm

#Simple regression of LogPriceA on LogPriceB to find residuals
model = sm.OLS(mergedData.LogPriceB,mergedData.LogPriceA)
model = model.fit()

print('ADF Test for the residual of OLS')
mergedData['Residual'] = mergedData.LogPriceB - model.params[0] * mergedData.LogPriceA

#If residual is stationary, it can be said that LogPriceA and LogPriceB are cointegrated.
ADFTest(mergedData['Residual'])
```

#### Pair trading

When comparing the theoretical price of B and the actual price of B, can you come up with a pairs trading model? Visualize and explain potential trading opportunities.

1) The spread is calculated such that it will be positive whenever calculated price of B is higher than market price. The trading system will use Bollinger Bands like bands on the spread so that the system enter into long position(Short B, Long A) when it touches lower band and short position(Long B, Short A) when it is on upper band. 2 standard deviation and 1000 lookback period are just values for the sake of example, which requires further optimisation.

2) The system never stops trading which means whenever a long position is closed, a short position will be opened vice versa.

3) With 80 trading signals, the system generates $14.057 profits per contract excluding fee, spread.

4) The returns are negatively skewed as visualized in the final histogram. This characteristic is the underlying reason of a huge drawdowns.



```python
mergedData['SpreadB'] = mergedData['tLogPriceB'] - mergedData['LogPriceB']

#Mean calculation for given lookback period(rate)
def getAverage(prices,rate):
    
    return prices.rolling(rate).mean()
            
#Calculate upper and lower band for spread - similar to Bollinger Bands
def getTradeBands(prices, rate=1000):
    sma = getAverage(prices, rate)
    std = prices.rolling(rate).std()
    bandUp = sma + std * 2 # Calculate top band
    bandDown = sma - std * 2 # Calculate bottom band
    return bandUp, bandDown

spreadPrices = mergedData['SpreadB']

bandUp, bandDown = getTradeBands(spreadPrices)

def pairsTradeStrategy(data, bandDown, bandUp):
    buyPrice = []
    sellPrice = []
    spreadSignal = []
    signal = 0

    for i in range(1,len(data)):
        if data[i-1] > bandDown[i-1] and data[i] < bandDown[i]:
            if signal != 1:
                buyPrice.append(data[i])
                sellPrice.append(np.nan)
                signal = 1
                spreadSignal.append(signal)
            else:
                buyPrice.append(np.nan)
                sellPrice.append(np.nan)
                spreadSignal.append(0)
        elif data[i-1] < bandUp[i-1] and data[i] > bandUp[i]:
            if signal != -1:
                buyPrice.append(np.nan)
                sellPrice.append(data[i])
                signal = -1
                spreadSignal.append(signal)
            else:
                buyPrice.append(np.nan)
                sellPrice.append(np.nan)
                spreadSignal.append(0)
        else:
            buyPrice.append(np.nan)
            sellPrice.append(np.nan)
            spreadSignal.append(0)
            
    return buyPrice, sellPrice, spreadSignal

buyPrice, sellPrice, spreadSignal = pairsTradeStrategy(mergedData['SpreadB'], bandDown, bandUp)
tradeFrame = mergedData[['PriceA','PriceB']].copy()
tradeFrame['Signal'] = pd.DataFrame(spreadSignal)

#Signal 1 implies long in Downband -> Short B, Long A
#Signal -1 implies short in UpBand -> Long B , Short A
position = 0
long = 0
short = 0
pnl = []
marginReq = []

def openLong(data,index,position,long):
    
    entryLongA = data['PriceA'].iloc[index]
    entryShortB = data['PriceB'].iloc[index]
    position = 1
    long = 1

    return entryLongA,entryShortB,long,position

def closeLong(data,index,position,long):
    
    exitLongA = data['PriceA'].iloc[index]
    exitShortB = data['PriceB'].iloc[index]
    position = 0
    long = 0
    
    return exitLongA,exitShortB,long,position

def openShort(data,index,position,short):
    
    entryShortA = data['PriceA'].iloc[index]
    entryShortA = data['PriceB'].iloc[index]
    position = 1
    short = 1
    
    return entryShortA,entryShortA,short,position

def closeShort(data,index,position,short):
    
    exitShortA = data['PriceA'].iloc[index]
    exitLongB = data['PriceB'].iloc[index]
    position = 0
    short = 0
    
    return exitShortA,exitLongB,short,position
    
for i in range(len(tradeFrame['Signal'])):
    
    if position == 0: #Open position loop
        if tradeFrame['Signal'].iloc[i] == 1: #Portfolio Long -> Short B, Long A
            entryLongA,entryShortB, position,long = openLong(tradeFrame,i,position,long)
            marginReq.append(entryLongA + entryShortB*1.5)
            continue
        elif tradeFrame['Signal'].iloc[i] == -1 : #Portfolio Short -> Long B, Short A
            entryShortA,entryLongB,position,short = openShort(tradeFrame,i,position,short)
            marginReq.append(entryShortA *1.5 + entryLongB)
            continue
    elif position == 1: #Close position loop
        if tradeFrame['Signal'].iloc[i] == -1 and long: #close long if exist
            exitLongA,exitShortB,position,long = closeLong(tradeFrame, i, position, long)
            profit = (exitLongA - entryLongA) + (entryShortB - exitShortB)
            pnl.append(round(profit,5))
            entryShortA,entryLongB,position,short = openShort(tradeFrame,i,position,short)
            marginReq.append(entryShortA *1.5 + entryLongB)
            continue
        elif tradeFrame['Signal'].iloc[i] == 1 and short: #close short if exist
            exitShortA,exitLongB,position,short = closeShort(tradeFrame,i,position,short)
            profit = (entryShortA - entryShortA) + (exitLongB - entryLongB)
            pnl.append(round(profit,5))
            entryLongA,entryShortB, position,long = openLong(tradeFrame,i,position,long)
            marginReq.append(entryLongA + entryShortB*1.5)
            continue
            

mergedData.SpreadB.plot(figsize=(30,15),label='Spread B',c='b')
plt.plot(bandUp,label='Up Spread', c='g')
plt.plot(bandDown,label='Down Spread', c='r')
plt.scatter(mergedData.iloc[1:].index, buyPrice, marker = '^', color = 'green', label = 'BUY', s = 200)
plt.scatter(mergedData.iloc[1:].index, sellPrice, marker = 'v', color = 'red', label = 'SELL', s = 200)
plt.legend(loc="upper left")
plt.show()

marginReq.pop(1)
totalsum = 0
totalcumsum = []
for i in pnl:
    
    totalsum = totalsum + i
    totalcumsum.append(totalsum)

tradeResults = pd.DataFrame(list(zip(pnl,totalcumsum,marginReq)),columns=['PnL','Cumulative','Margin'])
tradeResults['Returns'] = tradeResults['PnL']/tradeResults['Margin']*100
tradeResults.Cumulative.plot(figsize=(30,15),label='PnL Chart',c='b')
plt.bar(tradeResults.index, tradeResults.PnL, color ='maroon',width = 0.4)
plt.ylabel('Profit in $')
plt.show()


fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(1, 1, 1)
tradeResults['Returns'].hist(bins=50, ax=ax1)
ax1.set_xlabel('Return')
ax1.set_ylabel('Sample')
ax1.set_title('Return distribution')
plt.show()
```

#### Final Thoughts

What are potentially other things you ought to take into consideration when applying trading models to stock trading in the real world? How would you optimize the above model? On what markets do you think the above model would work well and on which markets do you think it would work poorly?

Potential Considerations for Trading System

1) Liquidity: The trader need to ensure that the market can provide enough liquidity for the trading signals. Even if the model is superb, non-filled positions cannot generate profits. 

2) Margin: "The market can stay irrational longer than you can stay solvent". There is no guarantee that spread will turn to mean, or zero, quickly to generate consistent profits. The bailout of LTCM in 1998 is a good example to show the effect of over-leveraged systems that aim to generate risk-free arbitrage profits.

3) Discrepancies on Model: The historical performance does not guarantee the future performance. Even after strictly optimized and backtested strategies can fail or stop working in the future. A consistent monitoring and a proper exit strategy are crucial.

4) Fee: Position fees can be fatal especially for HFT systems. A basic trading system that is designed in this assessment can generate more than 100 trading signals per day. Therefore, chasing extreme points to increase profitability can be a game-changer for this system. The trader should consider fees while accounting profitability on backtest.

5) Bid/Ask Spread: The system can be traded with a spread as small as 0.02 as seen from the chart above. Entering into a position with the best price as a maker is essential.

6) Pace: There might be other traders who are fighting for the same amount of ticks, therefore, a carefully designed, efficient and fast trading systems are required to implement these kind of strategies.

Optimization

While designing the trade system, there may be substantial amount of parameters to be changed. All model is based on some assumptions and all parameters are prone to optimization problems, for example, standard deviation threshold for spread, interpolation method for missing values etc. Moreover, while comparing different assets to find out the best pairs to trade, there is no guarantee or magical numbers of which the trader ensure that the best possible scenerio is found. Even so, optimization through back-test with past data is an indispensable part of designing trading system including walk-forward test with out-of-sample data.

Possible Markets

1) WTI vs Brent: With a long historical data, high-liquidity and volume, Oil market can provide the best possible opportunity for pairs trading.

2) Riding the Yield Curve: Another market that can provide high liquidity and volume as well as enormous amount of data to analyse.

3) Stocks that belong to the same sector: Possible liquidity problems in stock markets, fragile statistical relationship due to the operation of companies, i.e., an unexpected earning release performance may break the correlation and cointegration into pieces.
